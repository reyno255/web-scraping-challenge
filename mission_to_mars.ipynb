```python
import os
import pandas as pd
import requests

from bs4 import BeautifulSoup as bs
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

url = "https://mars.nasa.gov/news/?page=1&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest"

browser.visit(url)

html = browser.html
```

```python
# Create a Beautiful Soup object
soup = bs(html, 'html5lib')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all('li', class_="slide")
results
```

```python
dates = []
titles = []
paragraphs = []
links = []

# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        date = result.find('div', class_="list_date").text
        # Identify Title
        title = result.find('div', class_="content_title").text
        #Identify paragraphs
        paragraphs = result.find('div', class_="article_teaser_body").text
        # Identify and return link to listing
        link = result.a['href']
        
        dates.append(date)
        titles.append(title)
        links.append(link)
        # Print results only if title, price, and link are available
        if (title and link):
            print('-------------')
            print(date)
            print(title)
            print(paragraphs)
            print(link)
    except AttributeError as e:
        print(e)
        
df = pd.DataFrame(
                {'Date': dates, 'titles': titles,'paragraphs': paragraphs,'links':links}
                )
```
